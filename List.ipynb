1. Simple List

```python
fruits = ["apple", "banana", "cherry"]
print(fruits)

#A basic list of strings.
```

2. List with Mixed Data Types

```python
mixed = [25, "hello", 3.14, True]
print(mixed)
```

3. Accessing Elements

```python
numbers = [10, 20, 30, 40, 50]
print(numbers[0])   # First element
print(numbers[4])  # Last element
```

4. Updating Elements

```python
colors = ["red", "green", "blue"]
colors[0] = "yellow"
print(colors)
```

5. Appending to a List

```python
animals = ["dog", "cat"]
animals.append("lion")
print(animals)
```

6. Extending a List

```python
list1 = [1, 2]
list2 = [3, 4, 5]
list1.extend(list2)
print(list1)
```

7. Inserting into a List

```python
cars = ["BMW", "Audi", "Tesla"]
cars.insert(1, "Toyota")
print(cars)
```

8. Removing Elements

```python
languages = ["Python", "Java", "C++", "Python"]
languages.remove("Python")
print(languages)
```

9. List Slicing

```python
nums = [10, 20, 30, 40, 50, 60]
print(nums[1:4])   # from index 1 to 3
print(nums[:3])    # first 3 elements
print(nums[3:])    # elements from index 3 to end
```
